1. Which state has the highest crime in the year? 

```python
#Highest Crimes Committed

State_Crime = df.groupby('STATE')['NUMBER'].sum()
highest_crime = State_Crime.idxmax() 
highest_number = State_Crime.max() 
print(f'The state with the highest number is {highest_crime} with a total of {highest_number} crimes') 
```

```python
#Lowest Crimes Committed

lowest_crime = State_Crime.idxmin() 
lowest_number = State_Crime.min() 
print(f'The state with the lowest number is {lowest_crime} with a total of {lowest_number} crimes') 
```
