Fixing The Name Column

#Removing unwanted characters
df['Name'] = df['Name'].astype(str).str.replace('[!#?@$_]','',regex = True)
df['Name'] 

```python
df['Name'] = df['Name'].astype(str).str.replace('[^A-Za-z]', '', regex = True)
df['Name'] 
```

```python
#Capitalising the first letter of the Name
df['Name'] = df['Name'].astype(str).str.upper()
```

```python
df['Name'] = df['Name'].astype(str).str.lower()
#for lower casing
```

```python
df['Name'] = df['Name'].astype(str).str.title()

#For Title casing
```

```python
#Correcting the misspelt names
correct_names ={
    'Jhn' : 'John',
    'Peterrr' : 'Peter',
    'Sammm' : 'Sam',
    'Saarah' : 'Sarah',
    'Maryy' : 'Mary',
    'Annaa' : 'Anna',
    'Sarahh' : 'Sarah',
    '' : 'Unknown'
}
df['Name'] = df['Name'].replace(correct_names)
df['Name'] 
```

```python
df
```

Fixing the Gender Column

```python
df['Gender'] = df['Gender'].astype(str).str.title()
df['Gender'] 
```

```python
df['Gender'] = df['Gender'].astype(str).str.title().str.strip()
#Convert the gender column to title case and remove blank spaces
```

```python
Correct_Gender = {
    'F': 'Female',
    'M': 'Male',
    ' ': 'Unknown'
}
df['Gender'] = df['Gender'].replace(Correct_Gender)
df['Gender'] 
```

```python
df
```

```python
#Replace words to figures by using the dictionary format

Word_to_figure = {
    'Twenty': '20',
    'Thirty': '30',
    'forty': '40',
    'thirty-five':'35',
    '20 years': '20'
}
df['Age'] = df['Age'].replace(Word_to_figure)
df['Age'] 
```

```python
#Converting the age to numeric data type
df['Age'] = pd.to_numeric(df['Age'],errors = 'coerce')
df['Age']
```

```python
#Creating a variable called 'invalid_age' and converting the outliers(-1,-2) to null values

invalid_age = (df['Age'] < 0) | (df['Age'] > 100)
df.loc[invalid_age, 'Age'] = pd.NA
df['Age']
```

```python
df.loc[invalid_age, 'Age'] = None
df
```

```python
#filling the null with a mean value
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)
df['Age']
#Equating df['Age'] to df['Age'].fillna(mean_age) helps us get the new  df['Age']
```

```python
#fixing the date 
df['Date_Joined'] = pd.to_datetime(df['Date_Joined'],errors = 'coerce')
df['Date_Joined']
```

```python
#Using Dictionary format
Word2fig = {
     'Five Thousand': '5000',
     'seven thousand': '7000',
     'ten thousand': '10000',
     'One Million': '1000000',
     'four thousand':'4000',
     '400O':'4000',
     '4O00': '4000',
     '-2000':'2000'
 }
df['Salary'] = df['Salary'].replace(Word2fig)
```

```python
df['Salary'].astype(str).str.replace('[^0-9-.]', '', regex = True)
#Remove unwanted characters in Salary
```
